
## 1. Что такое динамическое программирование

**Идея:** разбить задачу на перекрывающиеся подзадачи и использовать их решения повторно.

Ключевые термины:
- **Состояние** (state): что мы уже «сделали» к этому моменту.
- **Переход** (transition/recurrence): как перейти из одного состояния в другое.
- **База** (base cases): граничные условия, с которых начинаем.
- **Порядок вычисления** (order): в каком порядке заполняем таблицу.
- **Память** (memory): как хранить результаты (таблица, массив).
- **Восстановление ответа** (traceback): как получить саму оптимальную структуру, а не только её стоимость.



## 2. «В коде»: Кузнечик — количество способов добраться до клетки `n`

Кузнечик сидит в клетке 0 и хочет допрыгать до клетки `n`. За один прыжок он может
перемещаться на 1 или 2 клетки вперёд.

**Состояние:** `dp[i]` — число способов добраться до клетки `i`.  
**База:** `dp[0] = 1` (стоим в нуле — один способ «уже здесь»).  
**Переход:** `dp[i] = dp[i-1] + dp[i-2]` для `i ≥ 1` (если шаги [1,2]).



Теперь в каждой клетке лежит целое число монет `coins[i]` (может быть отрицательным).
Кузнечик стартует в 0, на каждой посещённой клетке собирает монетки, и хочет добраться до `n` с **максимальной суммой**.

**Состояние:** `dp[i]` — максимальная сумма монет, которую можно собрать к моменту попадания в клетку `i`.  
**База:** `dp[0] = coins[0]`.  
**Переход:** `dp[i] = max(dp[i-k]) + coins[i]` по всем допустимым прыжкам `k`.  
Чтобы восстановить маршрут, храним `parent[i]` — из какой клетки пришли в оптимуме.
